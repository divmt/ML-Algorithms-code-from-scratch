# KMeans from scratch

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(data.drop(columns='class'),data['class'],shuffle=True,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
X_train_scaled=sc.fit_transform(X_train)
X_test_sclaed=sc.transform(X_test)
```

```python
def K_means(K,data,init=5):
    ''' Takes number of clusters and Data
     & returns cetroids'''
    data=np.array(data)
    final_centroids={}
    error=[]
    for n in range(init):
        centroids={}
        choice=np.random.choice(range(len(data)),K)              #choose k data points randomly 
        centroids={i:data[j] for i,j in zip(range(K),choice)}    #make them centroid (store in dictionary)
        
        # check the distance of each  data point from the cetroids and assign them cluster_id
        while True:  
    
            output=[]
            for i in data:
                lst=[]
                lst=[np.sqrt(np.sum((i-centroids[j])**2)) for j in centroids]
            
                output.append(lst.index(np.min(lst)))        #taking min distance index and adding it to list
    
            new=pd.DataFrame(data)
            new['group']=output
        
            
            new_centroids={i:new.groupby('group').mean().values[i] for i in np.unique(output)}
        
            check=np.sum([np.sqrt(np.sum((centroids[i]-new_centroids[i])**2)) for i in np.unique(output)])
        
            if(check>0.00001):
                centroids=new_centroids
            else:
                break
        final_centroids[n]=new_centroids
        error.append(np.sum([np.sqrt(np.sum((i[:-1]-new_centroids[i[-1]])**2)) for i in new.values]))
    return final_centroids[error.index(np.min(error))]
    
    
```

```python
centroids=K_means(3,X_train_scaled)
```

```python
def predict(centroids,data):
    data=np.array(data)
    output=[]
    for i in data:
        lst=[np.sqrt(np.sum((i-centroids[j])**2)) for j in centroids]
        output.append(lst.index(np.min(lst)))
    return output
```

```python
def Inertia(K,data):
    data=np.array(data)
    centroids=K_means(K,data)
    prediction=predict(centroids,data)
    return np.sum([np.sqrt(np.sum((i-centroids[j])**2)) for i,j in zip(data,prediction)])
    
```

```python
elbow=pd.Series([Inertia(i,X_train_scaled) for i in range(1,10)],index=range(1,10))
```

```python
plt.plot(elbow.index,elbow,marker='o')
plt.show()
```
